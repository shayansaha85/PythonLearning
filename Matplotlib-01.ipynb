```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,'r')
plt.xlabel('X VALUES')
plt.ylabel('Y VALUES')
plt.title('THIS IS A TEST PLOT')
plt.show()
```

## OOP

```python
figure = plt.figure()

axes = figure.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y, 'g')
axes.set_xlabel('TANMOY')
axes.set_ylabel('SUBHAM')
axes.set_title('SHAYAN')
```

```python
figure = plt.figure()

axes1 = figure.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = figure.add_axes([0.2, 0.5, 0.4, 0.3])

# BIG GRAPH
axes1.plot(x,y, 'g')
axes1.set_xlabel('TANMOY')
axes1.set_ylabel('SUBHAM')
axes1.set_title('SHAYAN')


# SMALL GRAPH

axes2.plot(y,x, 'b--')
axes2.set_xlabel('TANMOY1')
axes2.set_ylabel('SUBHAM1')
axes2.set_title('SHAYAN1')
```

```python
figure = plt.figure()
axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x**2, 'b--', label='x square plot')
axes.plot(x, x**3, 'gx-', label='x cube plot')
axes.legend(loc=2)
```

```python

# loc:
# 1 --> UR
# 2 --> UL
# 3 --> LL
# 4 --> LR
```

```python
figure = plt.figure()
axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x**2, 'b--', label='x square plot')
axes.plot(x, x**3, color='#EA7E96', label='x cube plot')
axes.legend(loc=2)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12,4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title('NORMAL')

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title('TIGHT')

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([1, 5])
axes[2].set_title('FLEXIBLE')
```
