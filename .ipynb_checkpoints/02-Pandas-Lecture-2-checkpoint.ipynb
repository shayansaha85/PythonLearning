```python
import pandas as pd

employeeData = pd.read_csv('SampleSheet.csv')

def incrementSalary(salary) :
    return round((salary + salary*0.3))

newSalary = list(employeeData['salary'].apply(incrementSalary))

newDataSet = {
    'firstname' : list(employeeData['firstname']),
    'lastname' :list(employeeData['lastname']),
    'age' :list(employeeData['age']),
    'company' :list(employeeData['company']),
    'salary' :list(employeeData['salary']),
    'notun_salary' : newSalary
}

new_employee_db = pd.DataFrame(newDataSet)



try: # jodi try er modhher code e error na thake, toile try execute hoibo
    new_employee_db.to_csv('NewEmployeeDB.csv', index=False)
    print("YOUR DATA IS SAVED SUCCESSFULLY")
    
except: # try e jekunu error thakle, except execute hoibo
    print("DATA SAVE ISSUE")
```

```python
import pandas as pd

employeeData = pd.read_csv('SampleSheet.csv')

def incrementSalary(salary) :
    return round((salary + salary*0.3))

newSalary = list(employeeData['salary'].apply(incrementSalary))

employeeData.insert(5, 'new_salary', newSalary)

try: # jodi try er modhher code e error na thake, toile try execute hoibo
    employeeData.to_csv('NewEmployeeDB___2.csv', index=False)
    print("YOUR DATA IS SAVED SUCCESSFULLY")
    
except: # try e jekunu error thakle, except execute hoibo
    print("DATA SAVE ISSUE")
```

# COMPLICATED TOPIC : JOIN

```python
import pandas as pd
```

```python
df_1 = pd.DataFrame({
    'colA' : ['a0','a1','a2','a3'],
    'colB' : ['b0','b1','b2','b3'],
    'colC' : ['c0','c1','c2','c3'],
    'colD' : ['d0','d1','d2','d3']
})


df_2 = pd.DataFrame({
    'colA' : ['a4','a5','a6','a7'],
    'colB' : ['b4','b5','b6','b7'],
    'colC' : ['c4','c5','c6','c7'],
    'colD' : ['d4','d5','d6','d7']
})


df_3 = pd.DataFrame({
    'colA' : ['a8','a9','a10','a11'],
    'colB' : ['b8','b9','b10','b11'],
    'colC' : ['c8','c9','c10','c11'],
    'colD' : ['d8','d9','d10','d11']
})
```

```python
df_1
```

```python
df_2
```

```python
df_3
```

```python
# Concatenation
```

```python
df_con_col = pd.concat([df_1, df_2, df_3])
df_con_col
```

```python
df_con_row = pd.concat([df_1, df_2, df_3], axis=1)
```

```python
df_con_row
```

```python
left_df = pd.DataFrame({
    'pk' : ['k1','k2','k3','k4'],
    'colA' : ['a1','a2','a3','a4'],
    'colB' : ['b1','b2','b3','b4']
})

right_df = pd.DataFrame({
    'pk' : ['k1','k2','k3','k4'],
    'colC' : ['c1','c2','c3','c4'],
    'colD' : ['d1','d2','d3','d4']
})
```

```python
left_df
```

```python
right_df
```

```python
pd.merge(left_df, right_df, how='inner', on='pk')
```

## One level harder

```python
l_df = pd.DataFrame({
    'pk1' : 'k0 k0 k1 k2'.split(),
    'pk2' : 'k0 k1 k0 k1'.split(),
    'colA' : 'a0 a1 a2 a3'.split(),
    'colB' : 'b0 b1 b2 b3'.split()
})

r_df = pd.DataFrame({
    'pk1' : 'k0 k1 k1 k2'.split(),
    'pk2' : 'k0 k0 k0 k0'.split(),
    'colC' : 'c0 c1 c2 c3'.split(),
    'colD' : 'd0 d1 d2 d3'.split()
})
```

```python
l_df
```

```python
r_df
```

```python
pd.merge(l_df, r_df, on=['pk1', 'pk2'])
```

```python
pd.merge(l_df, r_df, how='outer', on=['pk1', 'pk2'])
```

```python
pd.merge(l_df, r_df, how='left', on=['pk1', 'pk2'])
```

```python
pd.merge(l_df, r_df, how='right', on=['pk1', 'pk2'])
```

```python
l_df = pd.DataFrame({
    'colA' : 'a0 a1 a2'.split(),
    'colB' : 'b0 b1 b2'.split()},
    index='k1 k2 k3'.split())

r_df = pd.DataFrame({
    'colC' : 'c0 c1 c2'.split(),
    'colD' : 'd0 d1 d2'.split()},
    index='k2 k3 k4'.split()
)
```

```python
l_df
```

```python
r_df
```

```python
l_df.join(r_df)
```

```python
l_df.join(r_df, how='outer')
```
